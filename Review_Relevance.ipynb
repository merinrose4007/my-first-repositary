```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import ast
import pandas as pd
import numpy as np
import re

MODEL = SentenceTransformer('all-MiniLM-L6-v2')


def advanced_split(text):
    """
    Split on punctuation + connector words, but REMOVE connectors from the final segments.
    Good for outcome keyword matching.
    """
    if not isinstance(text, str) or not text.strip():
        return []

    # 1) split on punctuation
    chunks = re.split(r'[.!?;,]', text)

    # connector/contrast words used ONLY for splitting
    split_words = (
        r"(but|however|although|though|yet|nevertheless|despite|except|unfortunately|"
        r"on the contrary|instead|rather|and|or|nor|neither|either|because|"
        r"so|then|also|too|as well as)"
    )
    split_pattern = re.compile(rf"\b{split_words}\b", flags=re.IGNORECASE)

    segments = []

    for chunk in chunks:
        chunk = chunk.strip()
        if not chunk:
            continue

        # 2) split on connector words (do NOT keep delimiters)
        parts = split_pattern.split(chunk)

        # parts will include both text and matched delimiters.
        # We keep only the non-delimiter text parts.
        for part in parts:
            part = part.strip()
            if not part:
                continue
            # skip if the part itself is just a connector word
            if split_pattern.fullmatch(part):
                continue
            segments.append(part)
            # filter very short segment

    return segments

def split_for_claims(text):
    """
    Split text for claim analysis - only on major punctuation and contrast words
    """
    if not isinstance(text, str) or not text.strip():
        return []

    # Split only on punctuation that indicates complete thoughts
    punctuation_splits = re.split(r'[.!?;]', text)

    claim_segments = []
    claim_contrast = r'\b(?:but|however|although|though|yet|nevertheless|despite|except|unfortunately|on the contrary|instead|rather)\b'

    for chunk in punctuation_splits:
        if not chunk.strip():
            continue

        # Split on contrast words
        parts = re.split(f'({claim_contrast})', chunk, flags=re.IGNORECASE)

        current = []
        for part in parts:
            if re.fullmatch(claim_contrast, part, flags=re.IGNORECASE):
                if current:
                    claim_segments.append(' '.join(current))
                current = [part]
            else:
                if part.strip():
                    current.append(part)

        if current:
            claim_segments.append(' '.join(current))

    return [s.strip() for s in claim_segments if len(s.strip().split()) > 1]

from collections import defaultdict

def calculate_relevance_scores(product_df,
                               claim_hi=0.70,
                               claim_mid=0.40,
                               outcome_thr=0.45,
                               ratio_strong=0.20,
                               ratio_mid=0.10,
                               short_word_thresh=3):

    df = product_df.copy()

    try:
        claims = ast.literal_eval(df.iloc[0]['claim'])
        outcomes = ast.literal_eval(df.iloc[0]['outcomes of claim'])
    except:
        claims, outcomes = [], []

    if not claims:
        df["model_relevance"] = 0.0
        return df

    claims = [c for c in claims if isinstance(c, str) and c.strip()]
    outcomes = [o for o in outcomes if isinstance(o, str) and o.strip()]

    claim_emb = MODEL.encode(claims, normalize_embeddings=True)
    outcome_emb = MODEL.encode(outcomes, normalize_embeddings=True) if outcomes else None

    texts = df["text"].fillna("").astype(str).tolist()

    all_claim_segments = []
    all_outcome_segments = []
    claim_segment_indices = []
    outcome_segment_indices = []
    claim_segment_texts = []
    outcome_segment_texts = []

    for idx, review_text in enumerate(texts):

        claim_segments = split_for_claims(review_text)
        if not claim_segments:
            claim_segments = [review_text]

        all_claim_segments.extend(claim_segments)
        claim_segment_indices.extend([idx] * len(claim_segments))
        claim_segment_texts.extend(claim_segments)

        outcome_segments = advanced_split(review_text)
        if not outcome_segments:
            outcome_segments = [review_text]

        all_outcome_segments.extend(outcome_segments)
        outcome_segment_indices.extend([idx] * len(outcome_segments))
        outcome_segment_texts.extend(outcome_segments)

    claim_segment_emb = MODEL.encode(all_claim_segments, normalize_embeddings=True)
    outcome_segment_emb = MODEL.encode(all_outcome_segments, normalize_embeddings=True)

    segment_claim_sims = claim_segment_emb @ claim_emb.T
    segment_max_claim = segment_claim_sims.max(axis=1)

    if outcome_emb is not None and len(outcomes) > 0:
        segment_outcome_sims = outcome_segment_emb @ outcome_emb.T
        segment_outcome_matches = (segment_outcome_sims >= outcome_thr)
    else:
        segment_outcome_sims = None
        segment_outcome_matches = None

    # ✅ NEW: Build review → segment index maps ONCE
    claim_review_map = defaultdict(list)
    for seg_idx, review_idx in enumerate(claim_segment_indices):
        claim_review_map[review_idx].append(seg_idx)

    outcome_review_map = defaultdict(list)
    for seg_idx, review_idx in enumerate(outcome_segment_indices):
        outcome_review_map[review_idx].append(seg_idx)

    scores = []
    outcome_counts = []
    outcome_ratios = []
    review_outcome_details = []

    for review_idx in range(len(texts)):

        review_claim_indices = claim_review_map.get(review_idx, [])
        review_outcome_indices = outcome_review_map.get(review_idx, [])

        if not review_claim_indices:
            review_max_claim = 0.0
        else:
            review_max_claim = segment_max_claim[review_claim_indices].max()

        if segment_outcome_matches is not None and review_outcome_indices:
            review_outcome_matches = segment_outcome_matches[review_outcome_indices].any(axis=0)
            review_outcome_count = review_outcome_matches.sum()
            review_outcome_ratio = review_outcome_count / len(outcomes)

            outcome_details = []
            for seg_idx in review_outcome_indices:
                segment_text = outcome_segment_texts[seg_idx]
                seg_outcome_sims = segment_outcome_sims[seg_idx]

                for outcome_idx, outcome in enumerate(outcomes):
                    sim_score = float(seg_outcome_sims[outcome_idx])
                    is_match = sim_score >= outcome_thr

                    outcome_details.append({
                        'segment': segment_text,
                        'outcome': outcome,
                        'similarity': round(sim_score, 3),
                        'is_match': is_match
                    })

            unique_outcomes = {}
            for detail in outcome_details:
                outcome = detail['outcome']
                if outcome not in unique_outcomes or detail['similarity'] > unique_outcomes[outcome]['similarity']:
                    unique_outcomes[outcome] = detail

            review_outcome_details.append(list(unique_outcomes.values()))
        else:
            review_outcome_count = 0
            review_outcome_ratio = 0.0
            review_outcome_details.append([])

        word_count = len(texts[review_idx].split())
        mc = float(review_max_claim)
        ratio = float(review_outcome_ratio)

        if word_count < short_word_thresh:
            score = 0.4 if ratio > 0 else 0.2
        else:
            if mc > claim_hi:
                score = 1.0
            elif mc > claim_mid and ratio >= ratio_mid:
                score = 0.8
            elif ratio >= ratio_strong:
                score = 0.7
            elif ratio > 0:
                score = 0.5
            else:
                score = mc * 0.5

        scores.append(round(score, 2))
        outcome_counts.append(review_outcome_count)
        outcome_ratios.append(round(review_outcome_ratio, 3))

    df["model_relevance"] = scores
    df["outcome_ratio"] = outcome_ratios
    df["outcome_count"] = outcome_counts
    df["outcome_details"] = review_outcome_details

    return df
```

```python
def analyze_outcome_ratios(df):
    print("=== Outcome Ratio Distribution ===")
    print(df['outcome_ratio'].describe())

    bins = [0, 0.05, 0.1, 0.2, 0.3, 0.5, 1.0]
    df['ratio_bin'] = pd.cut(df['outcome_ratio'], bins)
    dist = df['ratio_bin'].value_counts().sort_index()

    for bin_range, count in dist.items():
        pct = count/len(df)*100
        print(f"{bin_range}: {count} reviews ({pct:.1f}%)")

    # Correlation with relevance
    corr = df['outcome_ratio'].corr(df['model_relevance'])
    print(f"\nCorrelation with relevance: {corr:.3f}")
```

```python
def rating_difference_metric(df):
    """
    Highly relevant reviews should have DIFFERENT ratings
    than irrelevant ones (not necessarily higher — just different)
    """
    results = []

    for product_id in df['parent_asin'].unique():
        product_df = df[df['parent_asin'] == product_id].copy()

        if len(product_df) < 10:
            continue

        top_avg = product_df.nlargest(3, 'model_relevance')['rating'].mean()
        bottom_avg = product_df.nsmallest(3, 'model_relevance')['rating'].mean()
        diff = abs(top_avg - bottom_avg)

        results.append({
            'product': product_id,
            'top_avg_rating': round(top_avg, 2),
            'bottom_avg_rating': round(bottom_avg, 2),
            'difference': round(diff, 2)
        })

    results_df = pd.DataFrame(results)
    print(f"Average rating difference: {results_df['difference'].mean():.2f}")
    print(f"Products with difference > 1.0: {(results_df['difference'] > 1.0).mean()*100:.0f}%")

    return results_df
```

```python
def score_distribution_metrics(df):
    """
    Check if scores are spread out, not all identical
    """
    results = []

    for product_id in df['parent_asin'].unique():
        product_df = df[df['parent_asin'] == product_id]
        scores = product_df['model_relevance']

        results.append({
            'product': product_id,
            'unique_scores': scores.nunique(),
            'std_dev': scores.std(),
            'percent_high': (scores >= 0.8).mean() * 100,
            'percent_medium': ((scores >= 0.4) & (scores < 0.8)).mean() * 100,
            'percent_low': (scores < 0.4).mean() * 100,
            'review_count': len(product_df)
        })

    results_df = pd.DataFrame(results)
    print("=== Score Distribution Summary ===")
    print(f"Avg unique scores per product: {results_df['unique_scores'].mean():.1f}")
    print(f"Avg std deviation: {results_df['std_dev'].mean():.3f}")
    print(f"Avg % high relevance: {results_df['percent_high'].mean():.1f}%")
    print(f"Avg % medium relevance: {results_df['percent_medium'].mean():.1f}%")
    print(f"Avg % low relevance: {results_df['percent_low'].mean():.1f}%")

    return results_df
```
